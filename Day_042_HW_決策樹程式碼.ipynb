```python
import numpy as np
import pandas as pd 
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score
import warnings
warnings.filterwarnings('ignore')
```

## 作業1

```python
# 讀取資料
wine = datasets.load_wine()
df_wine = pd.DataFrame(wine.data, columns=wine.feature_names)
target = pd.DataFrame(wine.target, columns=["Target"])
df = pd.concat([df_wine,target], axis=1)
df.head()
```

```python
# 切分訓練集/測試集
X_train, X_test, y_train, y_test = train_test_split(df_wine, target, test_size=0.3,  random_state=0)

# 建立模型
dtc1 = DecisionTreeClassifier(random_state=0)
dtc2 = DecisionTreeClassifier(max_depth=2, random_state=0)
dtc3 = DecisionTreeClassifier(min_samples_split=2, random_state=0)
dtc4 = DecisionTreeClassifier(min_samples_leaf=2, random_state=0)

# 訓練模型
dtc1.fit(X_train, y_train)
dtc2.fit(X_train, y_train)
dtc3.fit(X_train, y_train)
dtc4.fit(X_train, y_train)

# 預測測試集
y_pred1 = dtc1.predict(X_test)
y_pred2 = dtc2.predict(X_test)
y_pred3 = dtc3.predict(X_test)
y_pred4 = dtc4.predict(X_test)

# 預測值與實際值的準確度，使用 accuracy score
print("Model-1 Accuracy: %.2f" % accuracy_score(y_test, y_pred1))
print("Model-2 Accuracy: %.2f" % accuracy_score(y_test, y_pred2))
print("Model-3 Accuracy: %.2f" % accuracy_score(y_test, y_pred3))
print("Model-4 Accuracy: %.2f" % accuracy_score(y_test, y_pred4))
```

## 作業2

```python
# 讀取資料
boston = datasets.load_boston()
df_boston = pd.DataFrame(boston.data, columns=boston.feature_names)
target = pd.DataFrame(boston.target, columns=["Target"])
df = pd.concat([df_boston,target], axis=1)
df.head()
```

```python
# 切分訓練集/測試集
X_train, X_test, y_train, y_test = train_test_split(df_boston, target, test_size=0.3, random_state=0)

# 建立一個線性回歸模型
dtg = DecisionTreeRegressor(random_state=0)

# 將訓練資料丟進去模型訓練
dtg.fit(X_train, y_train)

# 將測試資料丟進模型得到預測結果
y_pred = dtg.predict(X_test)

# 預測值與實際值的差距，使用 MSE
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
```
