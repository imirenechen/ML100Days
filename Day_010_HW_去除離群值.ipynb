### [作業重點]

- 觀察將極端值資料直接刪除, 對於分布與迴歸分數的影響 (In[6], Out[6])

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df_train = pd.read_csv("data/house_train.csv.gz")

#np.log1p 是對原始的資料取Log運算, 有兩種效果: 可以把數字縮小、同時更貼近常態分布
train_Y = np.log1p(df_train["SalePrice"])
df = df_train.drop(["Id", "SalePrice"], axis=1)
df.head()
```

```python
# 只取int64, float64兩種"數值型"欄位, 存於 num_features 中
num_features = []
for dtype, feature in zip(df.dtypes, df.columns):
    if dtype == "float64" or dtype == "int64":
        num_features.append(feature)
print(f'{len(num_features)} Numeric Features : {num_features}')
```

```python
# 篩選數值型欄位
df = df[num_features]
df = df.fillna(-1)
df.head()
```

## 作業1
- 試著限制 '1樓地板面積(平方英尺)' (1stFlrSF) 欄位的上下限, 看看能否再進一步提高分數?

```python
# 顯示 1stFlrSF 與目標值的散佈圖
sns.regplot(df["1stFlrSF"], train_Y)
plt.show()

# 做線性迴歸, 觀察分數
MMEncoder = MinMaxScaler()
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 調整離群值
# 檢查分位數
print(np.percentile(df["1stFlrSF"], 3))
print(np.percentile(df["1stFlrSF"], 97))
```

```python
# clip:大於或小於閾值的值，以閾值取代
df["1stFlrSF"] = df["1stFlrSF"].clip(625.0,1981.6)
sns.regplot(df["1stFlrSF"], train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

## 作業2
- 續前題, 去除離群值有兩類方式: 捨棄離群值(刪除離群的資料) 以及調整離群值,  
請試著用同樣的上下限, 改為 '捨棄離群值' 的方法, 看看結果會變好還是變差? 並試著解釋原因。

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 捨棄離群值
keep_indexs = (df["1stFlrSF"] > 625.0) & (df["1stFlrSF"] < 1981.6)
df = df[keep_indexs]
train_Y = train_Y[keep_indexs]
sns.regplot(df["1stFlrSF"], train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```
