```python
train_data.describe()
```

```python
# 將三種不同類型的特徵分類 - int_features, float_features, object_features
int_features = []
float_features = []
object_features = []
for dtype, feature in zip(train_data.dtypes, train_data.columns):
    if dtype == 'float64':
        float_features.append(feature)
    elif dtype == 'int64':
        int_features.append(feature)
    else:
        object_features.append(feature)
print(f'{len(int_features)} Integer Features, {len(float_features)} Float Features, {len(object_features)} Object Features')
```

```python
# 計算三種不同類型的特徵的平均數、標準差
# 這裡只針對整數與浮點數進行(object在數值運算上會有問題)
train_data[int_features].describe()
```

```python
train_data[float_features].describe()
```

```python
# 拿其中一個欄位來計算平均數、標準差 ex. AMT_INCOME_TOTAL
print(train_data["AMT_INCOME_TOTAL"].mean())
print(train_data["AMT_INCOME_TOTAL"].std())
```

```python
# 畫直方圖 ex. OWN_CAR_AGE
plt.hist(train_data["OWN_CAR_AGE"])
plt.title("OWN_CAR_AGE")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# 畫長條圖 ex. TARGET
plt.bar(["0","1"], train_data["TARGET"].value_counts())
plt.show()
```

```python
# 當處理不為數值型態的特徵時，可以使用value_counts來計算其各唯一值數量
# plt.bar(train_data['CODE_GENDER'].unique(), train_data['CODE_GENDER'].value_counts())
# 上方寫法可能有誤, unique的排序不一定等於value_counts的排序
plt.bar(["F","M","XNA"], train_data["CODE_GENDER"].value_counts())
plt.show()
```
