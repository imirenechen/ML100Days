## 檢視與處理 Outliers
### 為何會有 outliers, 常見的 outlier 原因
- 未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999
- 可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本  

### [作業目標]
- 依照下列提示與引導, 以幾種不同的方式, 檢視可能的離群值  

### [作業重點]
- 從原始資料篩選可能的欄位, 看看那些欄位可能有離群值 (In[3], Out[3])
- 繪製目標值累積密度函數(ECDF)的圖形, 和常態分布的累積密度函數對比, 以確認是否有離群值的情形 (In[6], Out[6], In[7], Out[7])

```python
# Import 需要的套件
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("data/application_train.csv")
data.head()
```

### 請參考HomeCredit_columns_description.csv的欄位說明，觀察並列出三個你覺得可能有outlier的欄位並解釋可能的原因

```python
# 先篩選數值型的欄位
dtype_select = [np.dtype('int64'), np.dtype('float64')]
numeric_columns = list(data.columns[list(data.dtypes.isin(dtype_select))])

# 再把只有2值(通常是 0,1)的欄位去掉(只有兩值,就沒有所謂outlier)
numeric_columns = list(data[numeric_columns].columns[list(data[numeric_columns].apply(lambda x:len(x.unique())!=2))])
print("Numbers of remain columns: {}%".format(len(numeric_columns)/len(data.columns)*100))
```

```python
# 檢視這些欄位的數值範圍
for col in ["AMT_INCOME_TOTAL","REGION_POPULATION_RELATIVE","OBS_60_CNT_SOCIAL_CIRCLE"]:
    fig = plt.figure(figsize =(18, 5.5))
    ax1 = fig.add_subplot(121)
    ax2 = fig.add_subplot(122)
    data.boxplot(col,ax=ax2)
    data.hist(col,ax=ax1)
    plt.show()
```

### 從上面的圖檢查的結果，至少這三個欄位好像有點可疑
- AMT_INCOME_TOTAL
- REGION_POPULATION_RELATIVE
- OBS_60_CNT_SOCIAL_CIRCLE

### Hints: Emprical Cumulative Density Plot, [ECDF](https://zh.wikipedia.org/wiki/%E7%BB%8F%E9%AA%8C%E5%88%86%E5%B8%83%E5%87%BD%E6%95%B0), [ECDF with Python](https://stackoverflow.com/questions/14006520/ecdf-in-python-without-step-function)

```python
# 最大值離平均與中位數很遠
print(data["AMT_INCOME_TOTAL"].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)(應該就是常用的p-chart)
cdf = data["AMT_INCOME_TOTAL"].value_counts().sort_index().cumsum()

plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel("AMT_INCOME_TOTAL")
plt.ylabel("ECDF")
# 設定x和y軸的範圍
plt.xlim([cdf.index.min(), cdf.index.max()*1.05]) 
plt.ylim([-0.05,1.05])
plt.show()
```

```python
# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
plt.plot(np.log(list(cdf.index)), cdf/cdf.max())
plt.xlabel("AMT_INCOME_TOTAL(log-scale)")
plt.ylabel("ECDF")
plt.ylim([-0.05,1.05])
plt.show()
```

### 補充：Normal dist 常態分布的ECDF
- https://au.mathworks.com/help/examples/stats/win64/PlotEmpiricalCdfAndCompareWithSamplingDistributionExample_01.png

```python
# 最大值落在分布之外
print(data["REGION_POPULATION_RELATIVE"].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
cdf = data["REGION_POPULATION_RELATIVE"].value_counts().sort_index().cumsum()

plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel("REGION_POPULATION_RELATIVE")
plt.ylabel("ECDF")
plt.ylim([-0.05,1.05])
plt.show()

data["REGION_POPULATION_RELATIVE"].hist()
plt.show()

data["REGION_POPULATION_RELATIVE"].value_counts().sort_index()
# 就以這個欄位來說，雖然有資料掉在分布以外，也不算異常，僅代表這間公司在稍微熱鬧的地區有的據點較少，
# 導致 region population relative 在少的部分較為密集，但在大的部分較為疏漏
```

```python
# 最大值落在分布之外
print(data["OBS_60_CNT_SOCIAL_CIRCLE"].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
cdf = data["OBS_60_CNT_SOCIAL_CIRCLE"].value_counts().sort_index().cumsum()

plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel("OBS_60_CNT_SOCIAL_CIRCLE")
plt.ylabel("ECDF")
plt.xlim([cdf.index.min() * 0.95, cdf.index.max() * 1.05])
plt.ylim([-0.05,1.05])
plt.show()

data["OBS_60_CNT_SOCIAL_CIRCLE"].hist()
plt.show()

data["OBS_60_CNT_SOCIAL_CIRCLE"].value_counts().sort_index(ascending = False)
```

### 注意：當 histogram 畫出上面這種圖 (只出現一條，但是 x 軸延伸很長導致右邊有一大片空白時，代表右邊有值但是數量稀少。這時可以考慮用 value_counts 去找到這些數值

```python
# 把一些極端值暫時去掉，在繪製一次 Histogram
# 選擇 OBS_60_CNT_SOCIAL_CIRCLE 小於 20 的資料點繪製
loc_a = data["OBS_60_CNT_SOCIAL_CIRCLE"] <= 20
loc_b = "OBS_60_CNT_SOCIAL_CIRCLE"

data.loc[loc_a, loc_b].hist()
plt.show()
```
