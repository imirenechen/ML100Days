```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
plt.style.use('seaborn')
```

```python
# 建立一個 10 x 10, 數值分布自 -1.0 ~ 1.0 的矩陣
matrix = (np.random.random((10,10))-0.5)*2

#繪製 Heatmap
plt.figure(figsize=(10,10))
sns.heatmap(matrix, cmap='PuOr', annot=True)
plt.show()
```

```python
nrow = 1000
ncol = 3
# 建立一個 1000 x 3, 數值分布為 -1.0 ~ 1.0 的矩陣
matrix = (np.random.random((1000,3))-0.5)*2

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice).reset_index()

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False, 
                    hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])

# 上半部為 scatter, 對角線為 hist, 下半部為 density
grid.map_upper(plt.scatter, alpha=0.2)
grid.map_diag(sns.histplot)
grid.map_lower(sns.kdeplot, cmap=plt.cm.OrRd_r)
plt.show()
```

```python
nrow = 1000
ncol = 3
# 建立一個 1000 x 3, 數值分布為常態分佈的矩陣
matrix = np.random.randn(1000,3)

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice).reset_index()

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False,
                    hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])

# 上半部為 scatter, 對角線為 hist, 下半部為 density
grid.map_upper(plt.scatter, alpha=0.2)
grid.map_diag(sns.histplot)
grid.map_lower(sns.kdeplot, cmap=plt.cm.OrRd_r)
plt.show()
```
