```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
np.random.seed(1)
```

### 負相關

```python
# 隨機生成 1000 個介於 0~50 的數 x
x = np.random.randint(0, 50, 1000)

# 讓 y 與 x 呈負相關，再增加一些雜訊
y = 30 -2*x + np.random.normal(0, 10, 1000)

# 再用 numpy 裡的函數來計算相關係數
np.corrcoef(x, y)
```

```python
# 再看看正相關的 x,y 分布
plt.scatter(x, y)
plt.show()
```
